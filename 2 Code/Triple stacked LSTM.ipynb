#Functions for parsing midi files

##Define bounds

```python
lowerBound = 21
upperBound = 108
span = upperBound-lowerBound
```

##Function for note matrix to Midi file

```python
def noteStateMatrixToMidi(statematrix, name="example", span=span):
    statematrix = np.array(statematrix)
    if not len(statematrix.shape) == 3:
        statematrix = np.dstack((statematrix[:, :span], statematrix[:, span:]))
    statematrix = np.asarray(statematrix)
    pattern = midi.Pattern()
    track = midi.Track()
    pattern.append(track)
    
    span = upperBound-lowerBound
    tickscale = 45
    
    lastcmdtime = 0
    prevstate = [[0,0] for x in range(span)]
    for time, state in enumerate(statematrix + [prevstate[:]]):  
        offNotes = []
        onNotes = []
        for i in range(span):
            n = state[i]
            p = prevstate[i]
            if p[0] == 1:
                if n[0] == 0:
                    offNotes.append(i)
                elif n[1] == 1:
                    offNotes.append(i)
                    onNotes.append(i)
            elif n[0] == 1:
                onNotes.append(i)
        for note in offNotes:
            track.append(midi.NoteOffEvent(tick=(time-lastcmdtime)*tickscale, pitch=note+lowerBound))
            lastcmdtime = time
        for note in onNotes:
            track.append(midi.NoteOnEvent(tick=(time-lastcmdtime)*tickscale, velocity=40, pitch=note+lowerBound))
            lastcmdtime = time
            
        prevstate = state
    
    eot = midi.EndOfTrackEvent(tick=1)
    track.append(eot)

    midi.write_midifile("{}.midi".format(name), pattern) 
```

##Function for midi file to note matrix

```python
def midiToNoteStateMatrix(midifile, squash=True, span=span):
    pattern = midi.read_midifile(midifile)
    timeleft = [track[0].tick for track in pattern]
    posns = [0 for track in pattern]

    statematrix = []
    time = 0

    state = [[0,0] for x in range(span)]
    statematrix.append(state)
    condition = True
    while condition:
        if time % (pattern.resolution / 4) == (pattern.resolution / 8):
            # Crossed a note boundary. Create a new state, defaulting to holding notes
            oldstate = state
            state = [[oldstate[x][0],0] for x in range(span)]
            statematrix.append(state)
        
        for i in range(len(timeleft)): #For each track
            if not condition:
                break
            while timeleft[i] == 0:
                track = pattern[i]
                pos = posns[i]

                evt = track[pos]
                if isinstance(evt, midi.NoteEvent):
                    if (evt.pitch < lowerBound) or (evt.pitch >= upperBound):
                        pass
#                         print("Note {} at time {} out of bounds (ignoring)".format(evt.pitch, time))
                    else:
                        if isinstance(evt, midi.NoteOffEvent) or evt.velocity == 0:
                            state[evt.pitch-lowerBound] = [0, 0]
                        else:
                            state[evt.pitch-lowerBound] = [1, 1]
                elif isinstance(evt, midi.TimeSignatureEvent):
                    if evt.numerator not in (2, 4):
                        # We don't want to worry about non-4 time signatures. Bail early!
                        # print "Found time signature event {}. Bailing!".format(evt)
                        out =  statematrix
                        condition = False
                        break
                try:
                    timeleft[i] = track[pos + 1].tick
                    posns[i] += 1
                except IndexError:
                    timeleft[i] = None

            if timeleft[i] is not None:
                timeleft[i] -= 1

        if all(t is None for t in timeleft):
            break

        time += 1

    S = np.array(statematrix)
    statematrix = np.hstack((S[:, :, 0], S[:, :, 1]))
    statematrix = np.asarray(statematrix).tolist()
    return statematrix
```

#Parse midi

```python
input_ = midiToNoteStateMatrix('./songs/' + song_name)
input_ = np.array(input_)
print(input_.shape)

X_train_list = []
length = 110
for i in range(input_.shape[0] - length):
    X_train_list.append(input_[i:length+i])
X_train_org = np.array(X_train_list)
X_train_org = X_train_org[:3000, :,:]
print(X_train_org.shape)
X_train = X_train_org
```

```python
# Convert to list
encoded_snippets = [X_train[i,:,:] for i in range(X_train.shape[0])]
```

```python
NUM_SONGS = len(encoded_snippets)
print(str(NUM_SONGS) + " total snippets to learn from")
```

#Model Architecture

##Define hyperparameters

```python
input_size       = encoded_snippets[0].shape[1]   # The number of possible MIDI Notes
output_size      = input_size                  # Same as input size
hidden_size      = 128                         # Number of neurons in hidden layer

learning_rate    = 0.01 # Learning rate of the model
training_steps   = 2000  # Number of epochs during training
timesteps        = 100    # Length of song snippet -- this is what is fed into the model
```

##Define placeholders

```python
tf.reset_default_graph()

input_placeholder_shape = [None, timesteps, input_size] 
output_placeholder_shape = [None, output_size] 

input_vec  = tf.placeholder("float", input_placeholder_shape)  
output_vec = tf.placeholder("float", output_placeholder_shape)  
```

##Initialize weights

```python
# Initialize weights
weights = tf.Variable(tf.random_normal([hidden_size, output_size])) 
biases = tf.Variable(tf.random_normal([output_size]))
```

##Define RNN Cell - Triple stacked LSTM

```python
def RNN(input_vec, weights, biases):
    
    input_vec = tf.unstack(input_vec, timesteps, 1)

    cell1 = tf.nn.rnn_cell.BasicLSTMCell(hidden_size, forget_bias=1.0, state_is_tuple=True)
    cell2 = tf.nn.rnn_cell.BasicLSTMCell(hidden_size, forget_bias=1.0, state_is_tuple=True)
    cell3 = tf.nn.rnn_cell.BasicLSTMCell(hidden_size, forget_bias=1.0, state_is_tuple=True)

    stacked_cell = tf.nn.rnn_cell.MultiRNNCell([cell1, cell2, cell3], state_is_tuple=True)

    outputs, states = rnn.static_rnn(stacked_cell, input_vec, dtype=tf.float32)
    
    recurrent_net = (tf.matmul(outputs[-1], weights) + biases)
    
    prediction = tf.nn.sigmoid(recurrent_net)
    
    return recurrent_net, prediction
```

##Define loss

```python
warnings.simplefilter("ignore")
# logits and prediction
logits, prediction = RNN(input_vec, weights, biases)

# LOSS OPERATION:
loss_op = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(
    logits=logits, labels=output_vec))
```

##Define optimizer

```python
# TRAINING OPERATION:
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_op = optimizer.minimize(loss_op) 
```

##Define accuracy measurement logic

```python
# ACCURACY: We compute the accuracy in three steps.

# make notes 1 if prob > 0.3
pred_notes = tf.cast(tf.greater_equal(prediction, 0.3), tf.float32)

# take 'AND' condition for equality with actual note to be played - this makes sure all the notes are correct
correct_pred = tf.reduce_all(tf.equal(output_vec, pred_notes), 1)

# take mean of correctly predicted notes - gives % correct notes played
accuracy_op = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

#Model training

##Initialize variables

```python
init = tf.global_variables_initializer()
```

##Launch and run the session

```python
# 1) Launch the session
sess = tf.InteractiveSession()

# 2) Initialize the variables
sess.run(init)

# 3) Train!
display_step = 10 # how often to display progress
for step in range(training_steps):
    
    batch_x = X_train[:,:timesteps,:]
    batch_y = X_train[:,timesteps,:]

    # TRAINING: run the training operation with a feed_dict to fill in the placeholders
    '''TODO: Feed the training batch into the feed_dict.'''
    feed_dict = {
                    input_vec: batch_x,
                    output_vec: batch_y
                }
    sess.run(train_op, feed_dict=feed_dict)
    
    # DISPLAY METRICS
    if step % display_step == 0 or step == 1:
        # LOSS, ACCURACY: Compute the loss and accuracy by running both operations 
        loss, acc = sess.run([loss_op, accuracy_op], feed_dict=feed_dict)     
        suffix = "\nStep " + str(step) + ", Loss= " + \
                 "{:.4f}".format(loss) + ", Training Accuracy= " + \
                 "{:.3f}".format(acc)
        if acc >= 0.995:
          break
        print(suffix)
```

##Generation logic

###Boost

```python
# Use a random snippet as a seed for generating the new song.
bias = 60 #int(np.random.random()*25) + 20
GEN_SEED_RANDOMLY = True 
if GEN_SEED_RANDOMLY:
    ind = np.random.randint(NUM_SONGS)
else:
    ind = 297
    
gen_song = encoded_snippets[ind][:timesteps].tolist()

for i in range(len_gen_song):  
# boost
    if ( (np.array(gen_song[-100:]).sum() < np.array(encoded_snippets[ind][:timesteps].tolist()).sum() - 50)\
        & (len(gen_song[:-100]) >= 99) ):
        print('boost')
        ind = np.random.randint(NUM_SONGS)
        gen_song_boost = encoded_snippets[ind][:int(timesteps/4)].tolist()
        gen_song.extend(gen_song_boost)
# create seed
    # get the first 65 notes of seed
    seed_x = encoded_snippets[ind][:timesteps].tolist()[:bias]
    # get the remaining 35 from the previously generated snippet
    seed_new = gen_song[int(-timesteps+bias):]
    # append the two arrays
    seed_x.extend(seed_new)
    # convert to numpy array
    seed_x = np.array([seed_x])

# generate music!
    # input seed
    seed = {input_vec: seed_x}
    # generate new note
    pred_notes1 = sess.run(pred_notes, feed_dict = seed)
    # append notes to create song
    gen_song.append(np.array(pred_notes1[0]))

# print success %
    if ((i*100//len_gen_song == 25) & (i%10 == 0) ):
      print('25% complete')
    elif ((i*100//len_gen_song == 50) & (i%10 == 0) ):
      print('50% complete')
    elif ((i*100//len_gen_song == 75) & (i%10 == 0) ):
      print('75% complete')

print('100% complete')  
```
